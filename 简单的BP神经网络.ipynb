```python
import keras
from keras import utils
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = np.load('/home/kesci/input/TibetanMNIST5610/TibetanMNIST.npz')
```

```python
x_data = data['image'].reshape(17768, 784)
x_data = x_data/255
```

```python
y_data = utils.to_categorical(data['label'], 10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)
```

```python
# create model
model = keras.Sequential()
model.add(keras.layers.Dense(784, input_dim=784, kernel_initializer='normal', activation='tanh'))
model.add(keras.layers.Dense(512, kernel_initializer='normal', activation='tanh'))
model.add(keras.layers.Dense(512, kernel_initializer='normal', activation='tanh'))
model.add(keras.layers.Dense(10, kernel_initializer='normal', activation='softmax'))
```

```python
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X_train, y_train, epochs=20, batch_size=200, verbose=1)
```

```python
y_pre = model.predict(X_test)
```
